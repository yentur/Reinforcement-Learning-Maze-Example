```python
import numpy as np
import random
import time
import sys
if sys.version_info.major == 2:
    import Tkinter as tk
else:
    import tkinter as tk

UNIT = 40   # pixels
MAZE_H = 10  # grid height
MAZE_W = 10  # grid width


class Maze(tk.Tk, object):
    def __init__(self):
        super(Maze, self).__init__()
        self.action_space = ['u', 'd', 'l', 'r']
        self.n_actions = len(self.action_space)
        self.title('maze')
        self.geometry('{0}x{1}'.format(MAZE_H * UNIT, MAZE_H * UNIT))
        self._build_maze()

    def _build_maze(self):
        self.canvas = tk.Canvas(self, bg='white',
                           height=MAZE_H * UNIT,
                           width=MAZE_W * UNIT)

        for c in range(0, MAZE_W * UNIT, UNIT):
            x0, y0, x1, y1 = c, 0, c, MAZE_H * UNIT
            self.canvas.create_line(x0, y0, x1, y1)
        for r in range(0, MAZE_H * UNIT, UNIT):
            x0, y0, x1, y1 = 0, r, MAZE_W * UNIT, r
            self.canvas.create_line(x0, y0, x1, y1)

        origin = np.array([20, 20])
        
        point_center = origin + np.array([UNIT*4, UNIT*6])
        self.point = self.canvas.create_rectangle(
            point_center[0] - 15, point_center[1] - 15,
            point_center[0] + 15, point_center[1] + 15,
            fill='yellow')
        

        hell1_center = origin + np.array([UNIT*6, 0])
        self.hell1 = self.canvas.create_rectangle(
            hell1_center[0] - 15, hell1_center[1] - 15,
            hell1_center[0] + 15, hell1_center[1] + 15,
            fill='black')

        hell2_center = origin + np.array([UNIT*6, UNIT])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 1])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 3])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        hell2_center = origin + np.array([UNIT*6, UNIT * 4])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 5])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 6])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*6, UNIT * 7])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')

        hell2_center = origin + np.array([UNIT*6, UNIT * 9])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*4, UNIT * 5])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
    
        
        hell2_center = origin + np.array([UNIT*5, UNIT * 6])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*3, UNIT * 5])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*2, UNIT * 5])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*2, UNIT * 6])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black') 
        hell2_center = origin + np.array([UNIT*2, UNIT * 7])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*2, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*3, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        
        hell2_center = origin + np.array([UNIT*4, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black') 
        
        hell2_center = origin + np.array([UNIT*9, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black')
        hell2_center = origin + np.array([UNIT*8, UNIT * 8])
        self.hell2 = self.canvas.create_rectangle(
            hell2_center[0] - 15, hell2_center[1] - 15,
            hell2_center[0] + 15, hell2_center[1] + 15,
            fill='black') 

        oval_center = origin + UNIT * 9
        self.oval = self.canvas.create_oval(
            oval_center[0] - 15, oval_center[1] - 15,
            oval_center[0] + 15, oval_center[1] + 15,
            fill='green')

        self.rect = self.canvas.create_rectangle(
            origin[0] - 15, origin[1] - 15,
            origin[0] + 15, origin[1] + 15,
            fill='red')

        # pack all
        self.canvas.pack()

    def reset(self):
        self.update()
        time.sleep(0.5)
        self.canvas.delete(self.rect)
        origin = np.array([20, 20])
        self.rect = self.canvas.create_rectangle(
            origin[0] - 15, origin[1] - 15,
            origin[0] + 15, origin[1] + 15,
            fill='red')
        # return observation
        return self.canvas.coords(self.rect)

    def Move(self,x,y):
        self.canvas.move(self.rect, x, y)
    def render(self):
        time.sleep(0.1)
        self.update()
        


def animasyon(x,y,env):
    if x<0:
        for i in range(abs(x)):
            env.Move(-UNIT,0)
            env.render()
            
    else:
        for i in range(x):
            env.Move(UNIT,0)
            env.render()
            
    if y<0:
        for i in range(abs(y)):
            env.Move(0,-UNIT)
            env.render()
            
    else:
        for i in range(y):
            env.Move(0,UNIT)
            env.render()    
```

```python
class Field:
    def __init__(self, size, item_pickup, item_drop_off):
        self.size_x = size
        self.size_y = size
        self.odul1=False
        self.item_in_car = False
        self.item_position = item_pickup
        self.item_drop_off = item_drop_off
        self.position = (4,6)
    def get_number_of_states(self):
        return self.size_x*self.size_y*self.size_x*self.size_y*2
    def get_state(self):
        state = self.item_position[0]*(self.size_y*self.size_x*self.size_y*2)
        state += self.item_position[1]*(self.size_x*self.size_y*2)
        state += self.position[0] * (self.size_y * 2)
        state += self.position[1] * (2)
        if self.item_in_car:
            state += 1
        return state
    def move_driver(self, action):
        (x, y) = self.item_position
        
        if self.item_position == (6, 0):
            return -100, True
        elif self.item_position == (6, 1):
            return -100, True
        elif self.item_position == (6, 3):
            return -100, True
        elif self.item_position == (6, 4):
            return -100, True
        elif self.item_position == (6,5):
            return -100, True
        elif self.item_position == (6, 6):
            return -100, True
        elif self.item_position == (6, 7):
            return -100, True
        elif self.item_position == (6, 8):
            return -100, True  
        elif self.item_position == (6, 9):
            return -100, True         
        elif self.item_position == (4, 5):
            return -100, True
        elif self.item_position == (5, 6):
            return -100, True
        elif self.item_position == (3, 5):
            return -100, True
        elif self.item_position == (2, 5):
            return -100, True
        elif self.item_position == (2, 6):
            return -100, True
        elif self.item_position == (2, 7):
            return -100, True
        elif self.item_position == (2, 8):
            return -100, True
        elif self.item_position == (3, 8):
            return -100, True
        elif self.item_position == (4, 8):
            return -100, True
        elif self.item_position == (9, 8):
            return -100, True
        elif self.item_position == (8, 8):
            return -100, True


        
        if action == 0: 
            if y == 0:
                return -10, False
            else:
                self.item_position = (x, y-1)
                return -5, False
        elif action == 1: 
            if y == self.size_y - 1:
                return -10, False
            else:
                self.item_position = (x, y+1)
                return -5, False
        elif action == 2: 
            if x == self.size_x - 1:
                return -10, False
            else:
                self.item_position = (x+1, y)
                return -5, False
        elif action == 3:
            if x == 0:
                return -10, False
            else:
                self.item_position = (x-1, y)
                return -5, False

        elif action == 4:
            if self.item_in_car:
                return -10, False
            elif self.item_position != (4, 6):
                return -10, False
            else:
                self.item_in_car=True
                return 1, False

   
        elif action == 5: 
            if not self.item_in_car:
                return -10, False
            elif self.item_drop_off != (x, y):
                self.item_position = (x, y)
                return -20, False
            else:
                return 20, True
```

```python

def Reinforcement_Learning_Train(env,iterasyon=1000,tk=False):
    size = 10
    item_start = (0, 0)
    item_drop_off = (9, 9)
    old_position=(0,0)
    field = Field(size, item_start, item_drop_off,)
    states = field.get_number_of_states()
    actions = 6
    q_table = np.zeros((states, actions))
    alpha = 0.1
    gamma = 0.6
    epsilon = 0.1
    for i in range(iterasyon):
        field = Field(size, item_start, item_drop_off)
        done = False
        steps = 0
        while not done:
            state = field.get_state()
            if random.uniform(0, 1) < epsilon:
                action = random.randrange(0,6)
            else:
                action = np.argmax(q_table[state])
            reward, done = field.move_driver(action)
            next_state = field.get_state()
            old_value = q_table[state, action]
            next_max = np.max(q_table[next_state])
            new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max)
            q_table[state, action] = new_value
            steps += 1
            if tk==True and i%500==0:
                x=field.item_position[0]-old_position[0]
                y=field.item_position[1]-old_position[1]
                old_position=field.item_position
                animasyon(x,y,env)
                

    return q_table
                
            
def Reinforcement_Learning_Test(env,q_table):
    size = 10
    item_start = (0, 0)
    item_drop_off = (9, 9)
    old_position=(0,0)
    alpha = 0.1
    gamma = 0.6
    epsilon = 0.1
    field = Field(size, item_start, item_drop_off)
    done = False
    steps = 0
    while not done:
        state = field.get_state()
        if random.uniform(0, 1) < epsilon:
            action = random.randrange(0, 6)
        else:
            action = np.argmax(q_table[state])
        reward, done = field.move_driver(action)
        next_state = field.get_state()
        old_value = q_table[state, action]
        next_max = np.max(q_table[next_state])
        new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max)
        q_table[state, action] = new_value
        steps += 1
        x=field.item_position[0]-old_position[0]
        y=field.item_position[1]-old_position[1]
        old_position=field.item_position
        animasyon(x,y,env)
```

```python
env = Maze()
q_table=Reinforcement_Learning_Train(env,5000,True)
Reinforcement_Learning_Test(env,q_table)
env.mainloop()
```
